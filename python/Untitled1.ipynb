```python
import os
import numpy as np
import math as math
```

```python
def johannes(x0, y0, z0, vx0, vy0, vz0, t0=0, dt=0.001 ,nsteps=100):
    rp = math.sqrt( x0**2 + y0**2 + z0**2)
    G = 4*(math.pi**2)
    x = [x0] 
    y = [y0]
    z = [z0]
    t = [t0]
    vx = [vx0]
    vy = [vy0]
    vz = [vz0]
    for i in range(1, nsteps):
        Vxi = vx0 - ( (G) / rp**3 )*x0 *dt 
        xi = x0 + Vxi*dt
        Vyi = vy0 - ( (G) / rp**3 )*y0 *dt 
        yi = y0 + Vyi*dt
        Vzi = vz0 - ( (G) / rp**3 )*z0 *dt 
        zi = z0 + Vzi*dt
        x.append(xi)
        y.append(yi)
        z.append(zi)
        t.append(t0 + dt*i)
        x0 = xi
        y0 = yi
        z0 = zi
        vx0 = Vxi 
        vy0 = Vyi
        vz0 = Vzi
        
    return x, y, z, vx, vy, vz, t
```
