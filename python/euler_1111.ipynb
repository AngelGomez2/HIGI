```python
import os
import numpy as np 
import matplotlib.pyplot as plt
import math as math
from scipy.optimize import minimize
from scipy import stats
import sympy
```

```python
def euler(x0, v0, w0, t0=0, dt=0.01, nsteps=100):
  x=[x0]
  t=[t0]
  for i in range(1,nsteps):
    vi = v0 - w0**2*x0*dt
    xi = x0 + v0*dt
    x.append(xi)
    t.append(t0 + dt*i)
    x0 = xi
    v0 = vi
  return x,t
```

```python
x,t=euler(0,1,2,nsteps=1000)
plt.plot(t,x)
```

```python
def euler2(x0, v0, w0, t0=0, dt=0.01, nsteps=100):
  x=[x0]
  t=[t0]
  for i in range(1,nsteps):
    vi = v0 - (w0**2)*(np.sin(x0))*dt
    xi = x0 + v0*dt
    x.append(xi)
    t.append(t0 + dt*i)
    x0 = xi
    v0 = vi
  return x,t
```

```python
x,t=euler2(0,1,2,nsteps=1000)
plt.plot(t,x)
```
